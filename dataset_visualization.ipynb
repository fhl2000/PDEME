```python
import os
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1D_Advection: single file
## infomation
data_path = "/data1/zhouziyang/datasets/pdebench/1D/Advection/Train/1D_Advection_Sols_beta1.0.hdf5"
f = h5py.File(data_path, 'r')
print(f.keys())
for key in f.keys():
    print(f[key])
sample = f["tensor"][...][0]
f.close()
## visualization
X, Y = np.meshgrid(np.linspace(0,2,201), np.linspace(0,1,1024))
fig, ax = plt.subplots()
pc = ax.pcolormesh(X, Y, sample.T)
ax.set_xlabel("t")
ax.set_ylabel("x")
fig.colorbar(pc, ax=ax)
plt.show()
## save figure in eps format
fig_name = os.path.basename(data_path)
fig_name = os.path.splitext(fig_name)[0] + ".eps"
fig.savefig(fig_name, format="eps")
```

```python
# 2D Diffusion-Reaction: multi file
## information
data_path = "/data1/zhouziyang/datasets/pdebench/2D/diffusion-reaction/2D_diff-react_NA_NA.h5"
f = h5py.File(data_path, 'r')
print(f.keys())
print(len(f.keys()))
dataset = f['0000']
print(dataset.keys())
print(dataset['data'])
print(dataset['grid']['x']) # (-1, 1)
print(dataset['grid']['y']) # (-1, 1)
print(dataset['grid']['t']) # (0, 5)
u_sample = dataset['data'][..., 0]
v_sample = dataset['data'][..., 1]
f.close()
## visualization
X, Y = np.meshgrid(np.linspace(-1,1,128), np.linspace(-1,1,128))
fig, axs = plt.subplots(1, 3, figsize=(14,4), sharey=True)
for i, step in enumerate([0, 50, 100]):
    axs[i].pcolormesh(X, Y, u_sample[step].T)
    axs[i].set_title(f"t={2.5*i}")
    axs[i].set_xlabel("x")
    axs[i].set_ylabel("y")
fig.subplots_adjust(wspace=0.2)
plt.show()
## save figure in eps format
fig_name = os.path.basename(data_path)
fig_name = os.path.splitext(fig_name)[0] + ".eps"
fig.savefig(fig_name, format="eps")
```

```python
# 2D darcy flow: single file
## information
data_path = "/data1/zhouziyang/datasets/pdebench/2D/DarcyFlow/2D_DarcyFlow_beta1.0_Train.hdf5"
f = h5py.File(data_path, 'r')
print(f.keys())
keys = list(f.keys())
for key in keys:
    print(f[key])
input_sample = f["nu"][...][0]
label_sample = f["tensor"][...][0,0,...]
f.close()
## visualization
X, Y = np.meshgrid(np.linspace(0,1,128), np.linspace(0,1,128))
fig, axs = plt.subplots(1, 2, figsize=(9,4), sharey=True)
axs[0].pcolormesh(X, Y, input_sample.T)
axs[0].set_title(r"$\alpha(x)$")
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")
axs[1].pcolormesh(X, Y, label_sample.T)
axs[1].set_title(r"$u(x,y)$")
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
plt.subplots_adjust(wspace=0.1)
plt.show()
## save figure in eps format
fig_name = os.path.basename(data_path)
fig_name = os.path.splitext(fig_name)[0] + ".eps"
fig.savefig(fig_name, format="eps")
```

```python
# 2D Compressible NS: single file t \in (0,1) x \in (0,1)
## information
data_path = "/data1/zhouziyang/datasets/pdebench/2D/CFD/2D_Train_Rand/2D_CFD_Rand_M0.1_Eta0.1_Zeta0.1_periodic_128_Train.hdf5"
f = h5py.File(data_path, 'r')
print(f.keys())
for key in f.keys():
    print(f[key])
density_sample = f["density"][...][0]
t = f["t-coordinate"][...]
f.close()
## visualization
X, Y = np.meshgrid(np.linspace(0, 1, 128), np.linspace(0, 1, 128))
fig, ax = plt.subplots(1, 3, figsize=(14,4), sharey=True)
for i, step in enumerate([0, 10, 20]):
    pc = ax[i].pcolormesh(X, Y, density_sample[step].T)
    ax[i].set_title(f"t = {t[step]}")
    ax[i].set_xlabel("x")
    ax[i].set_ylabel("y")
fig.subplots_adjust(wspace=0.2)
plt.show()
## save figure in eps format
fig_name = os.path.basename(data_path)
fig_name = os.path.splitext(fig_name)[0] + ".eps"
fig.savefig(fig_name, format="eps")
```
