```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pde_names = ("1D Advection", "1D Diffusion-Reaction", "1D Burgers", 
             "1D Diffusion-Sorption", "2D DarcyFlow", "2D Shallow Water", 
             "1D Allen Cahn", "1D Cahn Hilliard")
RMSEs = {
    "PINNs": (0.3335, 0.0177, 0.0983, 0.0855, 0, 0.0125, 0, 0),
    "FNO": (0.0139, 0.0059, 0.0205, 0.0008, 0.0302, 0.0039, 0.0039, 0.0027),
    "UNet": (0.0628, 0.0526, 0.2292, 0.0569, 0.0196, 0.0959, 0.0344, 0.0537),
    "PINO": (0, 0, 0, 0, 0 ,0 ,0, 0),
    "DFVM": (0, 0, 0, 0, 0, 0, 0, 0)
}

x = np.arange(len(pde_names))*1.5  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained', figsize=(14, 4))

for attribute, measurement in RMSEs.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    multiplier += 1

ax.set_ylabel('RMSE')
ax.set_xticks(x + width*(multiplier//2), pde_names)
ax.legend(loc='upper right', ncols=5)
plt.show()
```
